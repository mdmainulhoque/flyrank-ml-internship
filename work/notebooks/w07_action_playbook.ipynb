## 1. Ranked actions + reason codes

*The queue: what to do first, and why, in words a human trusts.*

```python
import pandas as pd

playbook = pd.read_csv("../../work/outputs/baseline_action_score.csv")

playbook = (
    playbook
    .sort_values("baseline_score", ascending=False)
    .reset_index(drop=True)
)

playbook["Rank"] = playbook.index + 1

playbook = playbook[
    [
        "Rank",
        "content_id",
        "baseline_score",
        "action_label",
        "reason_code"
    ]
]

print("Top Ranked Content")
playbook.head(10)
```

## 3. Human review + the no-go list

*What a person must check before acting. What should never be automated.*

```python
print("=== Human Review Rules ===")

rules = [
    "Verify factual accuracy",
    "Check SEO intent",
    "Review brand voice",
    "Update references if necessary",
    "Approve manually before publishing"
]

for rule in rules:
    print("-", rule)

print("\n=== What Should NOT Be Automated ===")

no_go = [
    "Publishing content",
    "Deleting content",
    "Medical advice",
    "Legal advice",
    "Financial recommendations",
    "Final editorial approval"
]

for item in no_go:
    print("-", item)
```
